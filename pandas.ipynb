```python
import pandas as pd

# Substitua pelo nome correto do arquivo
recife = pd.read_csv('salarios_recife_outubro_2013.csv',
                 encoding='latin1',
                 converters={'R$SALARIO': lambda x: float(x.replace('R$', '').replace('.', '').replace(',', '.'))})

recife.head()
```

```python
orgao = recife.groupby('ORGAO')
orgao.count()
```

```python
print(orgao.count().head())
print('\n')
print(orgao.size().tail())
```

```python
print(orgao.sum() [5:10])
print('\n')
print(orgao['R$SALARIO'].mean() [5:10])
print('\n')
print(orgao['R$SALARIO'].median() [5:10])
```

```python
orgao['CARGO'].apply(lambda x: len(x.unique())).sort_values(ascending=False) [:5]
```

```python
def ranker(df):
  """Atribui uma classificação a cada funcionário com base no salário, sendo 1 o mais bem pago.
  Assume que os dados estão classificados com DESC."""
  df['orgao_rank'] = np.arange(len(df)) + 1
  return df
```

```python
recife.sort_values('R$SALARIO', ascending=False, inplace=True)
recife = recife.groupby('ORGAO').apply(ranker)
recife[recife.orgao_rank == 1].head(10)
```
